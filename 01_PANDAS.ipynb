```python
# FILTERING DATAFRAME ON A CONDITION / MASKING USING BOOLEAN SERIES 

data['city']=='Hyderabad'    #THIS FUNCTION WILL GIVE BOOLEAN OUTPUT 
```

```python
mask = data['city'] == 'Hyderabad'     #IT WILL GIVE THE ROWS OF THAT MATCHES WHICH WAS PLAYED IN HYDERABAD
data[mask]
```

```python
data[mask].shape[0]   # if we observe this output 49 indicates that, 49 matches were played at Hyderabad
```

```python
def get_city_match_count(city):      # WE CAN CREATE SUCH FUNCTIONS TO GET THE MATCH COUNTS AT OUR INPUT CITY
    mask=data['city']==city
    return data[mask].shape[0]
```

```python
get_city_match_count('Rajkot')
```
