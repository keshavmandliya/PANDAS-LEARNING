```python
#  GROUPBY ON IPL DATASET

import numpy as np
import pandas as pd
```

```python
delivery=pd.read_csv('deliveries.csv.zip')
```

```python
delivery.head(2)
```

```python
runs=delivery.groupby('batsman')
```

```python
runs.get_group('V Kohli')    # in this case the no. of rows shows the no. of balls played by virat kohli
```

```python
runs['batsman_runs'].sum()   # this will group all the batsmans list and sum all their runs infront of their names
```

```python
runs['batsman_runs'].sum().sort_values(ascending=False).head(5)   # in this way we can find out the top 5 run scorer of ipl
```

```python
runs['legbye_runs'].sum().sort_values(ascending=False)[['V Kohli']]   # this indicates that when V Kohli is on strike 39 legbye runs are scored
```

```python
mask=delivery['batsman_runs']==6
new_delivery=delivery[mask]
```

```python
new_delivery.shape[0]
```

```python
new_delivery.groupby('batsman')['batsman_runs'].count().sort_values(ascending=False).head(5)

# IN THIS WAY WE CAN FIND THE TOP 5 BATSMAN WHO SCORES MOST SIXES IN IPL HISTORY
```

```python
vk=delivery[delivery['batsman']=='V Kohli']
```

```python
vk.groupby('bowling_team')['batsman_runs'].sum().sort_values(ascending=False).head(10)
```

```python
def run_scored(batsman_name):
    vk=delivery[delivery['batsman']==batsman_name]
    return vk.groupby('bowling_team')['batsman_runs'].sum().sort_values(ascending=False).index[0]
```

```python
run_scored('V Kohli')
```

```python
run_scored('MS Dhoni')
```
